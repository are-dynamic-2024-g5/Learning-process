```python
import numpy as np
import random
from PIL import Image as im
from matplotlib import pyplot as plt
import time
```

```python
#Paramètres:

cooldown = 5 #Temps entre chaque pixel
cote_canvas = 10 #Taille du coté du canvas
matrice_modele1 = np.full((cote_canvas, cote_canvas), 1)
matrice_modele2 = np.full((cote_canvas, cote_canvas), 2)
canvas = np.zeros((cote_canvas, cote_canvas), dtype=np.uint8)
nb_agents = 90
ind_nb_agents = 0
ind_nb_coord = 1
ind_nb_couleur = 2
ind_nb_tecp = 3
```

```python
def plot_show(canvas):
    image = np.zeros((cote_canvas, cote_canvas, 3), dtype=np.uint8)
    color = {0: [255, 255, 255], 1: [255, 0, 0], 2: [0, 255, 0]}
    image[canvas == 0] = color[0]
    image[canvas == 1] = color[1]
    image[canvas == 2] = color[2]
    img = im.fromarray(image)
    img.save('output.png')
    img.show()
```

```python
def param_groupe(nb_agents, couleur):
    coordination = 1 / nb_agents #Plus on a d'agents moins on est coordonés.
    #correct_chance = nb_agents * cote_canvas + (cote_canvas - pixel_a_mettre)
    tecp = cooldown + (random.random() * coordination) #Temps entre chaque pixel
    #cooldown puis chance de poser qui augmente
    return [nb_agents, coordination, couleur, tecp]
    
    
```

```python
def trouve_bon_pixel(canvas, matrice_modele, couleur):
    for i in range(cote_canvas):
        for j in range(cote_canvas):
            if(matrice_modele[i][j] == couleur):
                return (i, j)
```

```python
def trouve_mauvais_pixel(canvas, matrice_modele, couleur):
    for i in range(cote_canvas):
        for j in range(cote_canvas):
            if(matrice_modele[i][j] != couleur):
                return (i, j)
```

```python
def coordonee(groupe, canvas, matrice_modele, nb_agents, couleur):
    #On prend un nombre au hasard, si on a une grande coordination alors y a une grande chance 
    #que le nomre pioché soit inférieur à coordination donc une grande chance de poser un pixel bon.
    if (int(random.random()) < groupe[ind_nb_coord]):  #Placer un bon pixel
        x, y = trouve_bon_pixel(canvas, matrice_modele, couleur)
    else:
        x, y = trouve_mauvais_pixel(canvas, matrice_modele, couleur)
    return (x, y)
```

```python
def simulation(g1, g2, matrice_modèle1, matrice_modèle2, canvas, nb_heures):
    """Fais affonter deux groupes"""
    nb_modifs_par_heure1 = 10000
    nb_modifs_par_heure2 = 20
    tmp1 = nb_modifs_par_heure1
    tmp2 = nb_modifs_par_heure2
    couleur_g1 = 1
    couleur_g2 = 2
    n = 0
    for i in range(nb_heures):
        print(n , canvas) #Affiche le canvas à chaque étape
        n += 1
        while((nb_modifs_par_heure1 > 0) or (nb_modifs_par_heure2 > 0)): #Tant qu'il reste des pixels à poser
            if(nb_modifs_par_heure1 > 0):#si il y a plus de pixels pour le groupe 1, on arrete de poser
                x1, y1 = coordonee(g1, canvas, matrice_modele1, g1[ind_nb_agents], 1)
                canvas[x1][y1] = couleur_g1
            if(nb_modifs_par_heure2 > 0): # pareil pour groupe 2
                x2, y2 = coordonee(g2, canvas, matrice_modele2, g2[ind_nb_agents], 2) 
                canvas[x2][y2] = couleur_g2
            nb_modifs_par_heure1 -= 1
            nb_modifs_par_heure2 -= 1
```

```python
simulation(param_groupe(500, 1), param_groupe(100, 2), matrice_modele1, matrice_modele2, canvas, 24)
```
